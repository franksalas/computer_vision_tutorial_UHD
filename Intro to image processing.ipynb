## Load,  display, edit  and save images

Check if we have the required libraries

````python
import cv2
import matplotlib.pyplot as plt

#%matplotlib inline
````

### How to load an image with OpenCV and Matplotlib
1. We create a variable and call `.imread()` method from cv2
2. we use matplotlib to display it on the notebook

````python
img = cv2.imread('images/RGB.png')

plt.imshow(img)
plt.show()
````

### Something is wrong!
OpenCV represents RGB images as multi-dimensional NumPy arrays…but in reverse order!

This means that images are actually represented in **BGR** order rather than **RGB**!

There’s an easy fix though.
All we need to do is convert the image from BGR to RGB with the method `cv2.cvtColor(img, cv2.COLOR_BGR2RGB)`

````python
img = cv2.imread('images/RGB.png')                # load image

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # convert image then show

plt.axis("off")   # remove x & y axis lines
plt.show()        # display image in jupyter notebook cell
````

## We can also convert the image to grayscale


````python
img = cv2.imread('images/RGB.png')                      # load image

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)            # create new variable with image converted to grayscale

plt.imshow(gray, cmap='gray', interpolation='bicubic')  # need it to display gray image correctly

plt.axis("off")  # remove x&y axis lines
plt.show()       # display image in jupyter notebook cell
````
