# CHALLENGE # 2 
FIND CATS!

EXAMPLE:

```python
# load image
img = cv2.imread('images/tubecat.jpg')
# convert from BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# create grayscale image
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# increase figure size
plt.figure(figsize=(10,10))

# load image to plot
plt.imshow(img)

# display image
plt.show()
```

```python
# cat detector cascade
detector = cv2.CascadeClassifier("haarcascades/haarcascade_frontalcatface_extended.xml")

# cat face location
rects = detector.detectMultiScale(gray, scaleFactor=1.05,
                                  minNeighbors=15, minSize=(75, 75))
# loop over to draw square
for (x,y, w, h) in rects:
    cv2.rectangle(img, (x,y), (x+ w, y + h), (0,255,0), 2)
    

plt.figure(figsize=(10,10))       # change the figure size to 20 by 10
plt.imshow(img)   # display 
plt.show()                        # display image onto jupyter notebook cell
```
